```python
# Titanic EDA Notebook

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")

titanic = sns.load_dataset('titanic')

print("\n--- Head of Dataset ---")
print(titanic.head())

print("\n--- Dataset Info ---")
titanic.info()

print("\n--- Summary Statistics ---")
print(titanic.describe())

print("\n--- Value Counts for 'sex' ---")
print(titanic['sex'].value_counts())

sns.pairplot(titanic.dropna(), hue='survived')
plt.suptitle('Pairplot by Survival', y=1.02)
plt.show()

plt.figure(figsize=(8, 5))
titanic['age'].hist(bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(x='survived', y='age', data=titanic)
plt.title('Age vs Survival')
plt.show()

sns.scatterplot(x='fare', y='age', hue='survived', data=titanic)
plt.title('Fare vs Age by Survival')
plt.show()


summary = """
Summary of Findings:
- Women had a much higher survival rate than men.
- Passengers in first class were more likely to survive.
- Younger passengers, especially children, had better survival rates.
- Fare and class were strongly correlated with survival.
- Some missing values exist in age and cabin columns.
"""
print(summary)
```
